```python
import matplotlib.pyplot as plt
```

```python
a=[]
b=[]
# y=0
# x=-50

for x in range(-50,50,1):
    y=x**2
    a.append(x)
    b.append(y)
    #x= x+1

fig= plt.figure()
axes=fig.add_subplot(111)
axes.plot(a,b)
plt.show()
```

```python
import numpy as np
x=np.arange(-50,50,0.1)
y=x**2

plt.ylabel('Y')
plt.xlabel('X')
plt.plot(x,y)
plt.show()
```
